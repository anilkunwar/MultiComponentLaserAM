```python
def uniform_time(power, scanspeed):
    POW, ScanSpeed = power, scanspeed
    csv_path = os.path.abspath(f'../../0_Moose_Files/{POW}W0{ScanSpeed}mps/exodus/')
    df = pd.read_csv(csv_path+'/HEA_Full.csv')
    index = int(df.index[df['dt'] == 2][0])   # df.index[df['dt'] != 2] ## to find index without dt=2
    # df = df[index:-1]
    df = df.iloc[index:-1].reset_index(drop=True)
    print(f'Power:{POW} Speed:{ScanSpeed} | dt=2 starts at ==>  {index}')
    return df
```
